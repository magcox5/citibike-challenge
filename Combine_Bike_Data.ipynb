```python
# Dependencies and Setup
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Files for 1 year's worth of data from monthly bike data 
bike_data = {}
bike_data[1] = "Data/202002-citibike-tripdata.csv"
bike_data[2] = "Data/202001-citibike-tripdata.csv"
bike_data[3] = "Data/201912-citibike-tripdata.csv"
bike_data[4] = "Data/201911-citibike-tripdata.csv"
bike_data[5] = "Data/201910-citibike-tripdata.csv"
bike_data[6] = "Data/201909-citibike-tripdata.csv"
bike_data[7] = "Data/201908-citibike-tripdata.csv"
bike_data[8] = "Data/201907-citibike-tripdata.csv"
bike_data[9] = "Data/201906-citibike-tripdata.csv"
bike_data[10] = "Data/201905-citibike-tripdata.csv"
bike_data[11]= "Data/201904-citibike-tripdata.csv"
bike_data[12] = "Data/201903-citibike-tripdata.csv"

# Read City bike data from monthly files
bike = {}
bike[1] = pd.read_csv(bike_data[1])
bike[2] = pd.read_csv(bike_data[2])
bike[3] = pd.read_csv(bike_data[3])
bike[4] = pd.read_csv(bike_data[4])
bike[5] = pd.read_csv(bike_data[5])
bike[6] = pd.read_csv(bike_data[6])
bike[7] = pd.read_csv(bike_data[7])
bike[8] = pd.read_csv(bike_data[8])
bike[9] = pd.read_csv(bike_data[9])
bike[10] = pd.read_csv(bike_data[10])
bike[11] = pd.read_csv(bike_data[11])
bike[12] = pd.read_csv(bike_data[12])
```

```python
bike_months = {}
bike_years = {}
for i in range(1,13):
    bike_months[i] = bike_data[i][9:11]
    bike_years[i] = bike_data[i][5:9]
    print(i, bike_data[i], bike_months[i], bike_years[i])
```

```python
# Add a month and year column to each dataframe
for i in range(1,13):
    bike[i]['Month'] = bike_months[i]
    bike[i]['Year'] = bike_years[i]
    bike[i]['time_minutes'] = round(bike[i]['tripduration']/60)
```

```python
# Combine all dataframes into 1 .csv
combined_bikes = pd.concat([bike[1], bike[2], bike[3], bike[4], bike[5], bike[6], bike[7], bike[8], bike[9], bike[10], bike[11], bike[12]])
combined_bikes
```
